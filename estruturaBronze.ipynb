```python
%sql
CREATE CATALOG IF NOT EXISTS assaia_catalog
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS assaia_catalog.bronze
```

```python
table_name = "assaia_catalog.bronze.dClientes"

if not spark.catalog.tableExists(table_name):

    spark.sql(f"""
-- =========================================================
-- 1) TABELAS  - BRONZE_CLIENTES
-- =========================================================
CREATE TABLE IF NOT EXISTS assaia_catalog.bronze.dClientes (

  _line                  bigint      COMMENT 'Identificaador Linha',
  _fivetran_synced       timestamp   COMMENT 'Identificaador ingestão',
  id	                   INT         COMMENT 'ID',
  nome_completo	         string      COMMENT 'Nome Completo',
  cpf	                   string      COMMENT 'CPF',
  genero	               string      COMMENT 'Gênero'

)
USING DELTA
COMMENT 'Bronze: eventos brutos do site (exemplo)'
TBLPROPERTIES (
  'delta.autoOptimize.optimizeWrite' = 'true',
  'delta.autoOptimize.autoCompact'   = 'true'
);
    """)

    print("Tabela criada.")
else:
    print("Tabela já existe.")
```
