```python
table_name = "assaia_catalog.gold.dimClientes"

if not spark.catalog.tableExists(table_name):

    spark.sql(f"""
-- =========================================================
-- 1) GOLD - dimClientes
-- =========================================================
CREATE TABLE IF NOT EXISTS assaia_catalog.gold.dimClientes (
  id_cliente              INT         NOT NULL COMMENT 'Chave de negócio do cliente',
  nome_completo           STRING      NOT NULL COMMENT 'Nome completo do cliente',
  cpf                     STRING      NOT NULL COMMENT 'CPF do cliente',
  genero                  STRING                  COMMENT 'Gênero do cliente',
  _dt_atualizacao         TIMESTAMP   NOT NULL COMMENT 'Data de atualização na gold',
  CONSTRAINT pk_gold_dimClientes PRIMARY KEY (id_cliente)
)
USING DELTA
COMMENT 'Gold: dimensão analítica de clientes'
TBLPROPERTIES (
  'delta.autoOptimize.optimizeWrite' = 'true',
  'delta.autoOptimize.autoCompact'   = 'true'
);
    """)

    spark.sql(f"""
ALTER TABLE assaia_catalog.gold.dimClientes
ADD CONSTRAINT chk_gold_dimClientes_nome
CHECK (trim(nome_completo) <> '');
    """)

    spark.sql(f"""
ALTER TABLE assaia_catalog.gold.dimClientes
ADD CONSTRAINT chk_gold_dimClientes_cpf
CHECK (length(regexp_replace(cpf, '[^0-9]', '')) = 11);
    """)

    print("Tabela criada.")
else:
    print("Tabela já existe.")
```
